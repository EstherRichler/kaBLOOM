```python
import os
from os import listdir
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plot image of chosen file

def show_image(date, location):
    
    #get filename
    filename = '/home/esther/shared/' + date + '_' + location + '.npy'
    
    #open file
    img = np.load(filename)
    
    #plot image
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15, 7))
    dat = ax.imshow(np.minimum(img[:, :, [3,2,1]] * 5, 1))
```
